#### 역함수 

- 어떤 함수의 입력/출력 관계와 정반대의 입출력 관계를 갖는 함수
- 원래의 함수에서 x축과 y축이 바뀐 것이므로, y=x가 나타내는 직선을 대칭축으로 하는 함수.

```python
def f1(x):
    return x**2
def f1inv(x):
    return np.sqrt(x)

x=np.linspace(0,3,300)
plt.figure(figsize=(10,6))
plt.plot(x,f1(x),"r-", label='함수 %f(x)=x^2$')
plt.plot(x,f1inv(x),"b-.", label='역함수 %f^{-1}(x)=\sqrt{x}$')
plt.plot(x,x,'g--')
plt.axis('equal')
plt.xlim(0,2)
plt.ylim(0,2)
plt.legend()
plt.title('역함수의 그래프')
plt.show()
```

#### 함수의 역함수를 구하는 과정

- 원래 함수의 x와 y를 바꿔줍니다.
- 바뀐 함수를 y에 대해서 푼 후, x로 정리합니다.

```python
# 과제3 역함수를 구하고 그래프로 표현하세요

def f(x):
    return (x**2+1)
def finv(x):
    return np.sqrt(x-1)

x= np.linspace(0,5,300)
plt.figure(figsize=(10,6))
plt.plot(x,f(x),"r-", label='함수 %f(x)=x^2$')
plt.plot(x,finv(x),"b-.", label='역함수 %f^{-1}(x)=\sqrt{x}$')
plt.plot(x,x,'g--')
plt.axis('equal')
plt.xlim(0,2)
plt.ylim(0,2)
plt.legend()
plt.title('역함수의 그래프')
plt.show()
```

#### 활성화함수

선형변환을 하면서 활성화 함수를 추가한다.

선형변환 과정에서 활성화 함수를 추가하여 인위적으로 비선형변환을 한다.

함수 예시: 나온 결과값이 0보다 작으면 0으로 처리, 0 이상이면 그 수로 처리(leru함수)


```python
# ReLU

xx = np.linspace(-10,10,100)
plt.figure(figsize=(3,2))
plt.plot(xx,np.maximum(xx,0)) # 만약 엑스가 음수면 최댓값은 0
plt.title('max(x,0) or ReLU')
plt.xlabel('$x$')
plt.ylabel('$ReLU(x)$')
```

#### 지수함수

- 정해진 어떤 값을 입력값으로 거듭제곱하여 출력을 구하는 함수
- 거듭 제곱을 할 때 아래에 있는 수를 밑이라고 하며, 밑을 오일러 수 e(약2.718)로하여 거듭제곱하는 함수를 지수함수라고 한다.
- $y=e^x$, y=exp(x) = exp x로 표시
    - e는 확률분포에서 많이 사용되는데, 정규분포에서 e의 지수 함수 형태인 exp(-x^2/2)
    - 로그함수의 밑으로 자연로그 함수(ln)을 사용하는데 이때 밑이 오일러상수 e이다.
    

```python
# 파이썬 코드로 하기
import math
print(math.e)
```

```python
# 로그함수의 밑이 e인 경우

x = 10
y = math.log(x)
y
```

```python
# 로그함수의 밑이 다른 수인 경우

x = 10
base = 10
y = math.log(x,base)
y
```

```python
print(np.exp(-10)) 
#자연상수 e의 -10승 값을 계산한 것입니다. 
#자연상수 e는 약 2.71828이고, 따라서 np.exp(-10)은 1/e의 10승인 약 0.0000454가 됩니다. 
#따라서, np.exp(-10)의 값은 약 0.0000454입니다.
print(np.exp(-1))
print(np.exp(-0.1))
print(np.exp(0))
print(np.exp(0.1))
print(np.exp(1))
print(np.exp(10))
```

```python
xx = np.linspace(-2,2,100)
yy = np.exp(xx) # y는 e의 x승
plt.figure(figsize=(5,3))
plt.title('지수함수')
plt.plot(xx,yy)
plt.axhline(1,c='r',ls='--')
plt.axhline(0,c='r',ls='--')
plt.axvline(0,c='r',ls='--')
plt.xlabel('$x$')
plt.ylabel('$\exp(x)$')
plt.show()
```
