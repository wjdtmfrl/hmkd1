```python
print("""
적정 수분 섭취량 계산 프로그램입니다.
아래를 진행하여 주시면 적정 수분 섭취량(ml)이 나옵니다.
""")

while 1: #실수 입력시까지 반복
    try: #예외처리위해 try 문법 사용
        weight = float(input("체중을 입력하세요 (kg): ")) #무게를 변수로. 실수 입력하게.
        if weight <= 0: 
            print("잘못된 체중 값입니다. 다시 입력해주세요.")
            continue #무게 변수가 0보다 작거나 같으면 재입력하게 컨티뉴 사용
        break #실수 입력시 다음으로.
    except ValueError: #수가 아닌 다른 값을 넣었을 시 오류는 예외처리
        print("잘못된 입력입니다. 숫자를 입력해주세요.")

while 1: #활동량 선택지 3가지 중 하나를 택할 때까지 반복
    try: #예외처리위해 try 문법 사용
        activity_level = int(input("활동량을 선택하세요 (1: 적음, 2: 보통, 3: 많음): ")) #활동레벨을 변수로. 정수만 입력하게.
        if activity_level not in [1, 2, 3]:#활동레벨을 1,2,3 중 택하지 않았을 경우
            print("잘못된 활동량 값입니다. 다시 입력해주세요.")
            continue #활동레벨을 1,2,3 중 택하지 않았을 경우 재입력하게.
        break #1,2,3 중 하나 입력시 다음으로.
    except ValueError: #수가 아닌 다른 값을 넣었을 시 오류는 예외처리
        print("잘못된 입력입니다. 숫자를 입력해주세요.")

while 1: #실수 입력시까지 반복
    try: #예외처리위해 try 문법 사용
        temperature = float(input("현재 체감 온도를 입력하세요 (℃): ")) #온도를 변수로. 실수 입력하게.
        if temperature <= -30: #온도를 -30 과 같거나 낮을 경우
            print("잘못된 체감 온도 값입니다. 다시 입력해주세요.")
            continue #온도를 -30 과 같거나 낮게 입력할 경우 재입력하게.
        break #온도 값을 정확히 작성할 시 다음으로.
    except ValueError: #수가 아닌 다른 값을 넣었을 시 오류는 예외처리
        print("잘못된 입력입니다. 숫자를 입력해주세요.")

#온도와 체중에 따른 수분량 구하는 공식을 if 조건문으로.
#recommended_intake 로 수분섭취량을 변수/결과값으로.
if temperature >= 28:
    recommended_intake = weight * 40
elif temperature >= 23:
    recommended_intake = weight * 35
elif temperature >= 20:
    recommended_intake = weight * 30
elif temperature >= 17:
    recommended_intake = weight * 25
elif temperature >= 10:
    recommended_intake = weight * 20
else:
    recommended_intake = weight * 15

if activity_level == 2:
    recommended_intake *= 1.2
elif activity_level == 3:
    recommended_intake *= 1.5
    
print("적정 수분 섭취량은 {:.1f} ml입니다.".format(recommended_intake))
#뽑아낸 결과값의 소수점 한자리까지만 나타내는 포맷으로 출력.

while 1:
    try:
        choice = input("계속 진행하려면 y를 입력하세요. 종료하려면 n을 입력하세요.: ")
        if choice.lower() == 'y': #문자열을 모두 소문자로 바꿔주는 lower 메소드 사용
            break
        elif choice.lower() == 'n':
            print("프로그램을 종료합니다.")
            exit() #프로그램을 완전히 종료하는.
        else:
            print("잘못된 입력입니다. y 또는 n을 입력해주세요.")
    except:
        continue
```
