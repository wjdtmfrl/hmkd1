```python
# 인자를 바꾸어 함수를 반복 호출하는 결과값 얻기 => map(적용하고싶은 함수,범위(리스트))
f= lambda x: x*x
args = [1,2,3,4,5] #이걸 다 적용하고 싶을 때, 반복호출: 이 함수에 대해서 각각의 원소들을 다 호출한다.

ret= map(f,args)

# print(type(ret)) => class = map.. 'map 객체'

print(ret)  #이렇게 나오면 list로 바꿔줘야 함
print(list(ret))

# 결과값이 map으로 타입이 바뀐거임 
```

#### python random 
1. random.random() : 0.0에서부터 1.0 사이의 실수(float)를 반환
2. random.uniform(a, b) : 인자로 들어온 a~b 사이의 실수(float)를 반환
3. randint(a, b) : 인자로 들어온 a, b 사이의 랜덤한 정수(int)를 반환
4. randrange(a, b), randrange(b) :  a <= x < b 의 범위 내에서의 랜덤한 정수(int)를 반환, randrange(b)는 0 <= x < b 의 범위 내에서의 랜덤한 정수(int)를 반환
5. random.choice(seq) : 인자로 받은 리스트, 튜플, 문자열, range 에서 무작위로 하나의 원소 반환
6. random.sample(seq or set, N) : 첫번째 인자로 받은 시퀀스데이터 or set에서 N개의 랜덤하고, unique하고, 순서상관없이 인자를 뽑아서 리스트로 만들어서 반환
7. random.shuffle(seq) : 데이터의 순서를 무작위로 랜덤하게 바꾸어 주는 함수

```python
# numbers 내부에 들어 있는 숫자가 몇 번 등장하는지 출력하는 코드를 작성하세요.

numbers = [3,6,3,5,8,2,9,7,4,3,6,3,5,8,2,9,7,4]
counter = {}  #사전 형식으로 나타내야 해서 {}로 받음.
for number in numbers:
    if number in counter:
        counter[number] = counter[number]+1
    else:
        counter[number] = 1

print(counter)
```

```python
# 반복문으로 팩토리얼을 구하는 사욯\ㅇ자 함수를작성하고 5!, 10!을 구하세요

def factorial(n):
    output=1
    for i in range(1,n+1):
        output *= i
    return output

print(factorial(5))
print(factorial(10))
```

```python
# Q. 재귀란 '자기자신을 호출하는 것'을 의미한다. 재귀함수를 이용해서 팩토리얼을 구하는 사용자 함수를 
# 작성하고 1!, 5!, 10!를 구하세요.
# factorial(n) = n * factorial(n-1) (n >= 1)

def factorial (n):
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)
    
print(factorial(5))
```

```python
# strip()함수를 사용하면 주어진 묹자열에서 양쪽 끝에 있는 공백과 \n 기호를 삭제
# re.sub() -> 정규표현식 문자열 치환
# title() 함수 첫 글자를 대문자로 변환

import re

states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
    "          'south   carolina##'", "'West virginia?'"]

def clean_strings (strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub('[!#?]','',value)
        value = value.title()
        result.append(value)
    return result

clean_strings(states)
```

```python
def getTextFreq(filename):
    with open(filename,'r')as f:
        text = f.read()
        fa = {}
        for c in text:
            if c in fa:
                fa[c] += 1
            else:
                fa[c] = 1
    return fa


ret = getTextFreq('mydata.txt')
ret = sorted(ret.items(),key=lambda x:x[1], reverse = True)
print(ret)

for c, freq in ret:
    if c == '\n':
        continue
    print('[%c] -> [%d]회 나타남' %(c,freq))
```

파이썬 if __name__ == "__main__" 의미

- 메인 함수의 선언, 시작을 의미합니다. 해당 코드 밑에 main 등의 함수 호출 코드를 작성해서 함수의 기능을 수행
- 직접 modex2.py 파일을 실행하는 경우에는 __name___ 변수에 __main__ 이라는 값이 할당되고 이 조건절이 참이 되어 if문 다음 문장들이 수행
- 모듈을 불러와서 사용하는 경우 __name__ 변수에는 모듈 이름이 저장되고 이 조건절이 거짓이 되어 if문 다음 문장들이 수행되지 않음
- if __name__ == "__main__"와 같이 조건문을 사용하여 터미널에서와 같이 직접 호출되어 사용될 때는 그 자체로 기능을 수행하고, 동시에 다른 모듈에서 필요한 함수 등을 제공

```python
with open('modex2.py','w') as f:
    f.write(
'''
def add(a,b):
    return a+b
if __name__=="__main__":    
    print(add(1,1))
''')
#     이거 실행하면 메인이 됨
```

#### [과제] 사칙연산 기능을 포함하는 모듈을 작성 후 임포트해서 사칙연산을 수행하세요.

- 모듈 cal4 작성 후 직접 실행하는 경우와 모듈 임포트하여 연산하는 두 가지 경우

```python
def cal4 (choice, *args):
    if choice == '+':
        result = 0
        for i in args:
            result += i
        return result
    
    elif choice == '-':
        result = 0
        for i in args:
            result -= i
        return result
        
    elif choice == '*':
        result = 1
        for i in args:
            result *= i
        return result
    
    elif choice == '/':
        result = 0
        for i,j in args:
            result = i/j
        return result
    
    else:
        print('다시입력')
        
cal4 ('-', 6,2)
```

#### 쓰레드 (Thread)

파이썬 프로그램은 기본적으로 하나의 쓰레드(Single Thread)에서 실행된다. 즉, 하나의 메인 쓰레드가 파이썬 코드를 순차적으로 실행한다. 코드를 병렬로 실행하기 위해서는 별도의 쓰레드(Subthread)를 생성해야 하는데, 파이썬에서 쓰레드를 생성하기 위해서는 threading 모듈 (High 레벨) 혹은 thread 모듈 (Low 레벨)을 사용할 수 있다. 일반적으로 쓰레드 처리를 위해서는 thread 모듈 위에서 구현된 threading 모듈을 사용하고 있으며, thread 모듈은 (deprecate 되어) 거의 사용하고 있지 않다.

```python
# 이 함수는 수행하는데 5초의 시간이 걸리는 함수다, 이 함수를 총 5번 반복 => 25초 소요.

import time

def long_task():
    for i in range(5):
        time.sleep(1)
        print('working: %s\n'%i)
        
print('Start')

for i in range(5):
    long_task()
    
print('End')
```

```python
# 파이썬에서 쓰레드를 실행하기 위해서는, threading 모듈의 threading.Thread() 함수를 호출하여 
# Thread 객체를 얻은 후 Thread 객체의 start() 메서드를 호출
# 쓰레드가 실행할 함수 (혹은 메서드)를 작성하고 그 함수명을 threading.Thread() 함수의 target 아큐먼트에 지정
# 프로그램을 수정하고 실행해 보면 25초 걸리던 작업이 5초 정도에 수행되는 것을 확인할 수 있다. 
# threading.Thread를 사용하여 만든 스레드 객체가 동시 작업을 가능하게 해 주기 때문임.


import time
import threading

def long_task():
    for i in range(5):
        time.sleep(1)
        print('working: %s\n'%i)

print('start')

threads = []
for i in range(5):
    t= threading.Thread(target=long_task)
    threads.append(t)
    
for t in threads:
    t.start()
    
for t in threads:
    t.join()   #join으로 스레드가 종료될때까지 기다린다.
    
print('end')
```

```python
# 뒤에서 3자리 씩 세서 콤마 넣기

num='123456789'
a=[]
k=0

for i in range(len(num),0,-1):
    a.append(num[i-1])
    k+=1
    if k==3 and i!=1:
        a.append(',')
        k=0
        
a.reverse()
print(''.join(a))
```

해당 코드는 문자열 "123456789"를 세 자리씩 끊어서 천 단위마다 쉼표를 추가하여 출력하는 것입니다.

우선, 빈 리스트 "a"와 변수 "k"를 초기화합니다.

반복문은 "for"로 시작하며, "range(len(num),0,-1)"은 변수 "i"가 "num"의 문자열 길이부터 1까지 역순으로 반복하도록 합니다.

반복문 내에서는 리스트 "a"에 "num" 문자열의 각 문자를 역순으로 추가합니다. 이때, 변수 "k"를 사용하여 3개의 문자열이 추가될 때마다 쉼표를 추가하고, 변수 "k"를 초기화합니다. 이 과정은 마지막 문자열을 제외한 모든 세 자리 문자열에 대해 반복됩니다.

마지막으로, 리스트 "a"를 역순으로 뒤집어서 순서를 원래대로 바꾸고, 문자열로 변환하여 출력합니다.

```python
num = input('아무 숫자를 입력하세요')
if num.isdigit():
    num = num[::-1]
    ret = ''
    for i,c in enumerate(num):
        i += 1
        if i !=len(num) and i%3 == 0:
            ret += (c+',')
        else:
            ret += c
    ret = ret[::-1]
    print(ret)
else:
    print(f'입력한 내용 {num}: 숫자가 아닙니다.')
```

#### 파이썬에서 -1은 다양한 방식으로 사용됩니다. 다음은 일부 예시입니다:

- 인덱스: 파이썬에서는 리스트, 문자열, 튜플 등의 시퀀스 자료형의 인덱스를 지정할 때 음수를 사용할 수 있습니다. 이때 -1은 시퀀스의 마지막 요소를 가리킵니다. 예를 들어, 리스트에서 my_list[-1]은 리스트의 마지막 요소를 반환합니다.

- 슬라이스: 마찬가지로, 시퀀스 자료형의 슬라이스를 지정할 때도 음수를 사용할 수 있습니다. 이때 -1은 시퀀스의 마지막 요소를 가리킵니다. 예를 들어, 리스트에서 my_list[:-1]은 마지막 요소를 제외한 모든 요소를 반환합니다.

- 반환 값: 일부 함수에서는 실패 또는 오류를 나타내는 값을 -1로 반환합니다. 예를 들어, 문자열에서 find() 메서드는 문자열에서 지정된 문자열을 찾을 수 없을 때 -1을 반환합니다.

- 비트 연산: 비트 연산에서는 -1이 모든 비트가 1인 이진수로 사용됩니다. 예를 들어, 0b11111111은 10진수로 255이며, -1과 같습니다.
